## 📦 Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set aesthetic style for the plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("viridis")
sns.set_context("notebook", font_scale=1.2)

# Ensure the output directory exists
output_dir = '../outputs/figures/'
os.makedirs(output_dir, exist_ok=True)

print('✅ Libraries imported successfully! Visual styles applied and output directory confirmed.')
```

---
## 📂 Step 2 — Load Data

```python
df = pd.read_csv('../data/processed/cleaned_covid.csv')
df.head()
```

---
## 🥇 Step 3 — Top 10 Countries Analysis

```python
def plot_top_10(data, column, title, color_palette, filename):
    """Helper function to plot Top 10 countries for a specific column."""
    top_10 = data.nlargest(10, column)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x=column, y='Country', data=top_10, palette=color_palette)
    
    plt.title(title, fontsize=16, pad=15)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Country', fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    
    # Save and show
    plt.savefig(f'{output_dir}{filename}', dpi=300)
    plt.show()
```

```python
plot_top_10(df, 'TotalCases', 'Top 10 Countries by Total Cases', 'Blues_r', '01_top10_cases.png')
```

```python
plot_top_10(df, 'TotalDeaths', 'Top 10 Countries by Total Deaths', 'Reds_r', '02_top10_deaths.png')
```

```python
plot_top_10(df, 'TotalRecovered', 'Top 10 Countries by Total Recoveries', 'Greens_r', '03_top10_recoveries.png')
```

## 🌍 Step 4 — Continent-level Comparison

```python
continent_group = df.groupby('Continent')[['TotalCases', 'TotalDeaths']].sum().reset_index()
continent_group = continent_group.sort_values(by='TotalCases', ascending=False)

labels = continent_group['Continent']
cases = continent_group['TotalCases']
deaths = continent_group['TotalDeaths']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 7))
rects1 = ax.bar(x - width/2, cases, width, label='Total Cases', color='#3498db')
rects2 = ax.bar(x + width/2, deaths, width, label='Total Deaths', color='#e74c3c')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Counts')
ax.set_title('Total Cases vs Total Deaths by Continent', pad=20, fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)

# Set y-axis to log scale because Cases are vastly higher than Deaths
ax.set_yscale('log')
ax.legend()

plt.tight_layout()
plt.savefig(f'{output_dir}04_continent_comparison.png', dpi=300)
plt.show()
```

---
## 💀 Step 5 — Case Fatality Rate (CFR) vs Recovery Rate
Let's look at the top 20 countries with the most cases, and plot their Fatality Rate.

```python
# Filter to top 20 countries by Total Cases to avoid noise from very small countries
top_20 = df.nlargest(20, 'TotalCases').sort_values('CaseFatalityRate', ascending=False)

plt.figure(figsize=(14, 8))
sns.barplot(x='CaseFatalityRate', y='Country', data=top_20, palette='flare')

plt.title('Case Fatality Rate (%) - Top 20 Most Affected Countries', fontsize=16, pad=15)
plt.xlabel('Case Fatality Rate (%)')
plt.ylabel('')

plt.tight_layout()
plt.savefig(f'{output_dir}05_case_fatality_rate.png', dpi=300)
plt.show()
```

---
## 🧪 Step 6 — Testing vs Total Cases Scatter Plot
Does doing more testing lead to finding more cases?

```python
plt.figure(figsize=(12, 8))

# Create scatter plot
sns.scatterplot(x='TotalTests', y='TotalCases', hue='Continent', size='Population', 
                sizes=(20, 800), alpha=0.7, palette='tab10', data=df)

# Use Log scales due to large outliers (like USA, India)
plt.xscale('log')
plt.yscale('log')

plt.title('Total Tests vs Total Cases (Log Scale)', fontsize=16, pad=15)
plt.xlabel('Total Tests (Log Range)')
plt.ylabel('Total Cases (Log Range)')

# Adjust legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig(f'{output_dir}06_tests_vs_cases.png', dpi=300)
plt.show()
```

---
## 📊 Step 7 — WHO Region breakdown (Pie Chart)

```python
who_cases = df.groupby('WHORegion')['TotalCases'].sum()

# Only show regions with > 1% of total cases, group others
threshold = who_cases.sum() * 0.01
mask = who_cases > threshold
tail = who_cases.loc[~mask].sum()
who_cases = who_cases.loc[mask]
if tail > 0:
    who_cases['Others'] = tail

plt.figure(figsize=(10, 10))
colors = sns.color_palette('pastel')[0:len(who_cases)]

plt.pie(who_cases, labels=who_cases.index, colors=colors, autopct='%1.1f%%', 
        startangle=140, pctdistance=0.85, wedgeprops={'edgecolor': 'white'})

# Draw circle for Donut Chart style
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Distribution of COVID-19 Cases by WHO Region', fontsize=16)
plt.tight_layout()
plt.savefig(f'{output_dir}07_who_region_donut.png', dpi=300)
plt.show()
```
